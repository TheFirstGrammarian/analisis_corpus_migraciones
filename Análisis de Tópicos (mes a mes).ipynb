```python
#Definicón de la función que devulve el topic_response
def topic_response(texto):
    topics_response =  meaningcloud.TopicsResponse(meaningcloud.TopicsRequest(license_key, txt=texto, lang='es', topicType='ec', otherparams=diccionario).sendReq())
    return(topics_response)
```

```python
# Definición de función para quitar '\n'
def clean(text):
    '''Get rid of some additional punctuation and non-sensical text that was missed the first time around.'''
    text = re.sub('\n', ' ', text)
    return text
```

## 4. Definición de funciones para los análisis mes a mes

```python
#Abrimos el corpus
responses = pickle.load(open("responses_topic_analysis", "rb"))
```

```python
#Definimos una función que devuelve el elemento más frecuente de una lista
def most_frequent(List): 
    occurence_count = Counter(List) 
    return occurence_count.most_common(1)[0][0] 
```

```python
#Definimos una función para presentar la información de las entidades en un df
def display_entidades(responses_en):
    entidades_en = []
    tipos_de_entidad = []
    relevancia = []
    numero_de_apariciones = []

    for response in responses_en:
        entidad_en = response.getEntities() #obtenemos las entidades de una noticia (lista con diccionario)
        for entidad in entidad_en:
            try:
                forma = entidad['official_form']
            except:
                forma = response.getTopicForm(entidad)
                
            tipo = [response.getTypeLastNode(response.getOntoType(entidad))]
            relevance = [int(entidad['relevance'])]
            num_ap = [len(entidad['variant_list'])]
            if forma not in entidades_en:
                entidades_en.append(forma)
                tipos_de_entidad.append(tipo)
                relevancia.append(relevance)
                numero_de_apariciones.append(num_ap)
            else: #que pasa si el elemento ya está en la lista??
                indice = entidades_en.index(forma)
                tipos_de_entidad[indice].append(tipo[0])
                relevancia[indice].append(relevance[0])         
                numero_de_apariciones[indice].append(num_ap[0])

    #entidades_en_2 = []
    tipos_de_entidad_2 = []
    relevancia_2 = []
    numero_de_apariciones_2 = []

    for tipo in tipos_de_entidad:
        tipo = most_frequent(tipo)
        tipos_de_entidad_2.append(tipo)
    for rel in relevancia:
        rel = sum(rel)/len(rel)
        relevancia_2.append(rel)
    for num in numero_de_apariciones:
        num = sum(num)
        numero_de_apariciones_2.append(num)

    ZippedList = list(zip(entidades_en, tipos_de_entidad_2, relevancia_2, numero_de_apariciones_2)) #Convertimos nuestras 4 listas en una lista de tuplas
    entidades_df = pd.DataFrame(ZippedList, columns = ['Entidad', 'Tipo', 'Relevancia', 'Número de apariciones'])#convertimos las tuplas en el dataframe y les damos un nombre a las columnas
    entidades_df.sort_values(by=['Número de apariciones'], inplace=True, ascending=False)
    return(entidades_df)   
```

```python
#Definimos una función para presentar la información de los conceptos en un df
def display_conceptos(responses_en):
    conceptos_en = []
    tipos_de_concepto = []
    relevancia = []
    numero_de_apariciones = []

    for response in responses_en:
        concepto_en = response.getConcepts() #obtenemos las conceptos de una noticia (lista con diccionario)
        for concepto in concepto_en:
            try:
                forma = concepto['official_form']
            except:
                forma = response.getTopicForm(concepto)
                
            tipo = [response.getTypeLastNode(response.getOntoType(concepto))]
            relevance = [int(concepto['relevance'])]
            num_ap = [len(concepto['variant_list'])]
            if forma not in conceptos_en:
                conceptos_en.append(forma)
                tipos_de_concepto.append(tipo)
                relevancia.append(relevance)
                numero_de_apariciones.append(num_ap)
            else: #que pasa si el elemento ya está en la lista??
                indice = conceptos_en.index(forma)
                tipos_de_concepto[indice].append(tipo[0])
                relevancia[indice].append(relevance[0])         
                numero_de_apariciones[indice].append(num_ap[0])

    tipos_de_concepto_2 = []
    relevancia_2 = []
    numero_de_apariciones_2 = []

    for tipo in tipos_de_concepto:
        tipo = most_frequent(tipo)
        tipos_de_concepto_2.append(tipo)
    for rel in relevancia:
        rel = sum(rel)/len(rel)
        relevancia_2.append(rel)
    for num in numero_de_apariciones:
        num = sum(num)
        numero_de_apariciones_2.append(num)

    ZippedList = list(zip(conceptos_en, tipos_de_concepto_2, relevancia_2, numero_de_apariciones_2)) #Convertimos nuestras 4 listas en una lista de tuplas
    conceptos_df = pd.DataFrame(ZippedList, columns = ['concepto', 'Tipo', 'Relevancia', 'Número de apariciones'])#convertimos las tuplas en el dataframe y les damos un nombre a las columnas
    conceptos_df.sort_values(by=['Número de apariciones'], inplace=True, ascending=False)
    return(conceptos_df)
```
