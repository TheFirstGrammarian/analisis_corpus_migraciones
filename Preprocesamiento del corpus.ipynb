## Pre-procesamiento del corpus

Convertimos nuestras noticias (guardadas en distintas carpetas por mes, a su vez contenidas en una carpeta "corpus" más general), en un dataframe de panda. El mismo es guardado con pickle para ser utilizado en nuestros análisis.

```python
#Declaramos las librerías a utilizar
import pickle
import pandas as pd
import os
import codecs
```

```python
#definimos una función que toma el mes y convierte las noticias contenidas en esa carpeta en un dataframe de pandas
def pd_mes(mes):
    archivos_del_corpus = './corpus/' + mes + '/'
    corpus = [] #lista en donde se guardan los textos
    doc_names = [] #lista en donde se guardan los títulos de las noticias
    meses = [] #lista en donde se guarda el mes
    
    for doc in os.listdir(archivos_del_corpus): #Leemos los documentos en la carpeta del mes
        archivo = codecs.open(archivos_del_corpus + doc, 'r', encoding = 'utf-8')
        doc = archivo.read()
        corpus.append(doc)
        doc_names.append(os.path.basename(archivo.name)) #esta línea devuelve el nombre del archivo
        meses.append(mes)

    ZippedList = list(zip(meses, doc_names, corpus)) #Convertimos nuestras tres listas en una lista de tuplas
    df =pd.DataFrame(ZippedList, columns = ['mes', 'Título', 'Noticia'])#convertimos las tuplas en el dataframe y les damos un nombre a las columnas
    return(df)#devolvemos el data frame
```

```python
#por cada carpeta del corpus (que contiene las noticias de un mes en particular), creamos un dataframe y luego los unimos todos
meses = ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo', 'Junio']
noticias = pd.DataFrame() #declaramos un dataframe vacío
for mes in meses:
    df_mes = pd_mes(mes) #creamos un dataframe de cada mes
    noticias = noticias.append(df_mes, ignore_index=True)#vamos concatenando los dataframes de cada mes
```

```python
#Guardamos nuestras noticias en un dataframe
noticias.to_pickle("corpus.pkl")
```
